Problem 1: Sample Mean and Variance

```python
## create an array ##
x= np.array([10, 16, 26, 12, 17, 22, 14, 12, 21, 16])
print(x)
```

```python
## sample mean & sample variance ##
l=len(x)
mn=np.sum(x)/l
diff=(x-mn)
var=np.sum(diff)/(l-1)
print("sample mean:",mn)
print("sample variance:",var)
```

Problem 2: Scoring a Regression Model

```python
## Function for SSE ##
def find_see(true_y,pred_y):
  l=len(true_y)
  diff=true_y-pred_y
  output=np.sum(diff*diff)
  return output
```

```python
## SSE for Model 1 & Model 2 ##
true=np.array([22.1, 17.9, 16.5, 14.3, 19.8, 23.7, 22.0, 18.4, 25.7, 19.2])
pred1=np.array([21.4, 16.7, 17.9, 12.1, 22.1, 25.1, 21.7, 19.3, 23.4, 19.9])
pred2=np.array([20.7, 18.1, 16.9, 13.6, 21.9, 24.8, 20.3, 21.1, 24.8, 18.4])
sse_1=find_see(true,pred1)
sse_2=find_see(true,pred2)
print("Model 1 SSE:",sse_1)
print("Model 2 SSE:",sse_2)
```

Problem 4: Classification Report

```python
## function for classification report ##
def classification_report(true_y,pred_y):
  classes=np.unique(true_y)
  a=classes[1]
  b=classes[0]
  t=np.count_nonzero(true_y)
  p=np.count_nonzero(pred_y==a)
  output=p/t
  TP=np.count_nonzero((true_y==a) & (pred_y==a) & (true_y==pred_y))
  FP=np.count_nonzero((true_y==b) & (pred_y==a))
  TN=np.count_nonzero((true_y==b) & (pred_y==b) & (true_y==pred_y))
  FN=np.count_nonzero((true_y==a) & (pred_y==b))
  PP=TP/(TP+FP)
  PR=TP/(TP+FN)
  NP=TN/(TN+FN)
  NR=TN/(TN+FP)
  print('Possitive class:',a)
  print('Negative class:',b,"\n")
  print('accuracy:',output)
  print("Positive Precision:",PP)
  print("Positive Recall:",PR)
  print("Negative Precision:",NP)
  print("Negative Recall:",NR)
```

```python
## classification report for detecting the presence of a disease in patients ##
true_diag=np.array(['P', 'P', 'N', 'N', 'P', 'N', 'N', 'N', 'P', 'N', 'N', 'N', 'N', 'P', 'P', 'N', 'N', 'N', 'N', 'N'])
pred_diag=np.array(['N', 'P', 'N', 'P', 'P', 'N', 'P', 'N', 'P', 'N', 'N', 'N', 'P', 'P', 'P', 'N', 'N', 'N', 'P', 'N'])
classification_report(true_diag,pred_diag)
```

```python
## Classification report of label images of dogs and cats ##
true_labels=np.array(['dog', 'dog', 'cat', 'dog', 'cat', 'cat', 'cat', 'dog', 'cat', 'cat', 'dog', 'cat',
'cat', 'dog', 'dog', 'dog', 'dog', 'cat', 'cat', 'cat', 'dog', 'dog', 'cat', 'cat'])
pred_labels=np.array(['dog', 'dog', 'cat', 'dog', 'cat', 'dog', 'cat', 'dog', 'cat', 'cat', 'dog', 'cat',
'cat', 'dog', 'cat', 'dog', 'dog', 'cat', 'dog', 'cat', 'dog', 'dog', 'cat', 'cat'])
classification_report(true_labels,pred_labels)
```

Problem 6: Stochastic Linear Relationships

```python
## Numpy array ## 
x_vals=np.random.normal(10,2,200)
errors=np.random.normal(0,1.2,200)
y_vals=5.1+0.9*x_vals+errors
print(y_vals)
```

```python
## Scatter Plot ##
width_height_2 = (8, 6)
plt.figure(figsize=width_height_2)
plt.scatter(x_vals,y_vals,s=60,alpha=0.8,color='green',edgecolor='black')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title("scatter plot")
plt.show()
```

```python
## Correlation ##
mu_x=np.mean(x_vals)
mu_y=np.mean(y_vals)
diff_x=(x_vals-mu_x)
diff_y=(y_vals-mu_y)
upper=np.sum(diff_x*diff_y)
lower=np.sqrt((np.sum(diff_x*diff_x))*(np.sum(diff_y*diff_y)))
r=upper/lower
print('Correlation between X and Y :',r)
```
