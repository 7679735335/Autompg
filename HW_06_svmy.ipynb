Part 4: Load Auto MPG Dataset

```python
## load the Autompg dataset ##
auto = pd.read_csv('autompg.txt', sep='\t')
auto.head(10)
```

```python
## shape of data ##
auto.shape
```

```python
## mean for all variables in data ##
auto.mean()
```

Part 5: Regional Counts and Means

```python
## unique regions ##
regions=np.unique(auto['region'])
print(regions)
```

```python
## number of car manufactured in various regions ##
mask=auto.region==regions[0]
Asia_auto=auto[mask]
l1=len(Asia_auto)
mask=auto.region==regions[1]
Eur_auto=auto[mask]
l2=len(Eur_auto)
mask=auto.region==regions[2]
Usa_auto=auto[mask]
l3=len(Usa_auto)
print("Number of cars manufactured in Asia:",l1)
print("Number of cars manufactured in Europe:",l2)
print("Number of cars manufactured in USA:",l3)
```

```python
asia_mean=Asia_auto.mean()
eur_mean=Eur_auto.mean()
usa_mean=Usa_auto.mean()
```

```python
## table of mpg, eur and wt for different regions ##
numpy_data = np.array([asia_mean, eur_mean,usa_mean])
mean_df = pd.DataFrame(data=numpy_data, index=["asia", "eur",'usa'], columns=["mpg", "eur",'wt'])
print(mean_df)
```

Part 6: Average Weight and MPG by Region

```python
## bar plot for average MPG by region ##
color1=['red','skyblue','green']
width_height_2 = (8, 4)
plt.figure(figsize=width_height_2)
plt.bar(mean_df.index,mean_df.mpg,color=color1)
plt.title("Average MPG by Region")
plt.xlabel("Region")
plt.ylabel('Average MPG')
plt.tight_layout()
plt.show()
```

```python
## bar plot for average Weight by region ##
color1=['red','skyblue','green']
width_height_2 = (8, 4)
plt.figure(figsize=width_height_2)
plt.bar(mean_df.index,mean_df.wt,color=color1)
plt.title("Average Weight by Region")
plt.xlabel("Region")
plt.ylabel('Average Weight in Pounds')
plt.tight_layout()
plt.show()
```

Part 7: Relationship between Weight and Miles Per Gallon

```python
## Scatter plot of Weight Vs MPG for different Regions ##
for index in range(len(regions)):
  mask=auto.region==regions[index]
  sc=auto[mask]
  width_height_2 = (12, 4)
  plt.figure(figsize=width_height_2)
  plt.xlim=([1200,5000])
  plt.ylim=([0,50])
  plt.subplot(1,3,index+1)
  plt.scatter(sc.wt,sc.mpg,alpha=0.8,edgecolors='B')
  plt.xlabel("Weight in Pounds")
  plt.ylabel("Miles Per Gallon")
  plt.title("Weight vs MPG (%s)"%regions[index])
  plt.tight_layout()
  plt.show()
```

Part 8: Cylinder Distribution by Region

```python
## unique values of Cyl in auto dataset ##
cyl_values=np.unique(auto.cyl)
print(cyl_values)
```

```python
## Cross tab between region and cyl ##
cyl_counts_by_region=pd.crosstab(auto.cyl,auto.region)
cyl_counts_by_region.head()
```

```python
## Proportion table between Region and Cyl ##
cyl_props_by_region=cyl_counts_by_region.loc[:,regions].div(cyl_counts_by_region.sum(axis=0), axis=1)
print(cyl_props_by_region)
```

```python
## stacked bar chart ##
bar_bottoms = np.cumsum(cyl_props_by_region) - cyl_props_by_region
color2=['red','blue','green','skyblue','B','maroon']
p=[0]
width_height_2 = (6, 4)
plt.figure(figsize=width_height_2)
for index in cyl_values:
  hi=cyl_props_by_region.loc[index,:]
  plt.bar(x=['asia','eur','usa'],height=hi,bottom=bar_bottoms.loc[index,:],color=color2[index-3])
  plt.legend([index])
plt.xlabel("Region")
plt.ylabel('Proportion')
plt.title("Distribution of Cylinder Numbers by Region")
plt.show()
```
